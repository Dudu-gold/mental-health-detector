Import libraries and load the dataset

```python
import pandas as pd
import re
import nltk
import string
import emoji

df = pd.read_csv('../data/mental_health_social_media_dataset.csv')
df.head()
```

Download and import **stopwords** and **tokenizer** tool

```python
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('wordnet', quiet=True)     
nltk.download('omw-1.4', quiet=True) 
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

Create **stopwords** and **lemmatizer** instance/object

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

Create **cleaning** function

```python

def clean_text(text, max_length=30):
    
    text = emoji.demojize(text)

    text = text.lower().strip()

    text = re.sub(r'[^a-z\s]', '', re.sub(r'\b([a-z])\1+\b', '', text))

    tokens = word_tokenize(text)

    tokens = [word for word in tokens if word.isalpha()]

    tokens = [word for word in tokens if word not in stop_words]

    tokens = [lemmatizer.lemmatize(word, pos='v') for word in tokens]

    tokens = [word for word in tokens if len(word) <= max_length]

    return " ".join(tokens)
```

Apply the cleaning function to dataset

```python
df['clean_text'] = df['text'].apply(clean_text)
```

Check result

```python
df[['text', 'clean_text']].head()
```

Save the **cleaned** and **preprocessed** dataset

```python
df.to_csv('mental_health_dataset_cleaned.csv', index=False)
```
